```python
counts=[]
for file in glob('./density_data/*count*'):
    counts.append(cis.read_data(file, 'poc_mask_num_points'))
total_counts = make_from_cube(squeeze(sum(counts)))
```

```python
sums=[]
for file in glob('./density_data/*sum*'):
    sums.append(cis.read_data(file, 'poc_mask'))
total_sum = make_from_cube(squeeze(sum(sums)))
```

```python
import matplotlib.ticker as mticker
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import iris.plot as iplt

extents = {'California':[-155,-95,5,55], 'Peru':[-115,-55,-40,10], 'Namibia':[-25,35,-40,10]}

fig, axes = plt.subplots(1, 3, figsize=(15,10), subplot_kw=dict(projection=ccrs.PlateCarree()))#, dpi=300)

for place, ax in zip(extents.keys(), axes):
    plt.sca(ax)
    p=iplt.pcolormesh(total_sum.copy(np.ma.masked_where(total_counts.data<10000, total_sum.data/total_counts.data)), vmin=0, vmax=30, cmap='Blues')
#     p=iplt.pcolormesh(total_sum/total_counts, vmin=0, vmax=30, cmap='Blues')
    c=qplt.contour(acverage_sc, cmap='Oranges', vmin=0.0, vmax=100.0)
    ax.set_extent(extents[place])
    ax.set_title(place)
    ax.coastlines()
    
    gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True)
    gl.xlines = False
    gl.ylines = False
    gl.xlabels_top = False
    gl.ylabels_right = False
    gl.ylocator = mticker.AutoLocator()
    
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    
# fig.subplots_adjust(bottom=0.8)
# cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
# fig.colorbar(p, cax=cbar_ax, )
fig.colorbar(p, ax=axes.ravel().tolist(), orientation='horizontal', label='POC density (%)', shrink=0.8, anchor=(0.,4.0))
fig.colorbar(c, ax=axes.ravel().tolist(), orientation='vertical', label='ISCCP Sc amount (%)', shrink=0.5)
# plt.tight_layout()
plt.savefig('POC_density.png')
```
